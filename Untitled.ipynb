## Importing Libraries


```python

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

## Importing the Data

```python
#Importing the data
water = pd.read_csv("C:/Users/hp/Downloads/water_potability.csv")
```

## Data Preprocessing

```python
print(water.shape)
```

```python
#visualizing the top 5 rows of the dataset
water.head()
```

```python
#visualizing the bottom 5 rows of the dataset
water.tail()
```

```python
#getting column information 
water.info()
```

```python
#missing values
water.isnull().sum()
```

```python
water['ph'] = water['ph'].fillna(water['ph'].mean())
water['Sulfate'] = water['Sulfate'].fillna(water['Sulfate'].min())
water['Trihalomethanes'] = water['Trihalomethanes'].fillna(water['Trihalomethanes'].mean())
```

```python
water.isnull().sum()
```

## Exploratory Data Analysis

```python
#Finding Correlation between the column
cor = water.corr()
cor.style.background_gradient(cmap='coolwarm')
```

```python
#Count of Potable and Non-Potable
water.Potability.value_counts()
sns.countplot(water.Potability,label="count")
plt.show()
```

```python
water['ph'] = water['ph'].round(decimals = 1)
def water_type(x):
    if x>7:
        y='Alkaline'
    elif x<7:
        y='Acidic'
    elif x==7:
        y='Pure Water'
    return y
        
water['Type'] = water['ph'].apply(lambda x: water_type(x))
```

```python
plt.figure(figsize=(7,5))

water['Type'].value_counts().plot(kind='pie',labels = ['','','','','',''], autopct='%1.1f%%',
                                  colors = ['red','yellow','blue'],  textprops = {"fontsize":17})

plt.legend(labels=['Alkaline', 'Pure Water','Acidic'])
plt.show()
```

```python
water_p = pd.get_dummies(water, columns = ['Type'])
water_p.head()
```

```python
water_Potable= water_p.drop(['Type_Acidic','Type_Alkaline','Type_Pure Water'],axis=1)
```

```python
X=water_Potable.drop(['Potability'],axis=1)
y=water_Potable['Potability']
```

## Train and Test Split

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.3, random_state=49)
```

## Machine Learning Algorithm

## Logistic Regression

```python

LR_Model = LogisticRegression()
LR_Model.fit(X_train,y_train)
predict_rf=LR_Model.predict(X_test)
accuracy = accuracy_score(y_test,predict_rf)
recall = recall_score(y_test, predict_rf, average="weighted")
precision = precision_score(y_test, predict_rf, average="weighted")
#f1_score = f1_score(y_test, predictions, average="micro")
f1_score = 2 * ((precision * recall) / (precision + recall))

print("LogisticRegression")
print("Accuracy: "+"{:.2f}%".format(accuracy * 100))
print("Recall:"+"{:.2f}%".format(recall * 100))
print("Precision: "+"{:.2f}%".format(precision * 100))
print("F1 Score:"+"{:.2f}%".format(f1_score * 100))
```

## Decision Tree

```python

Decision_Tree = DecisionTreeClassifier()
Decision_Tree.fit(X_train, y_train)
#Testing the Model on test set
predict_dt=Decision_Tree.predict(X_test)
acccuracy = accuracy_score(y_test, predict_dt)
recall = recall_score(y_test, predict_dt, average="weighted")
precision = precision_score(y_test, predict_dt, average="weighted")
f1_score = 2 * ((precision * recall) / (precision + recall))

print("Decision Tree Results")
print("Accuracy:"+"{:.2f}%".format(acccuracy * 100))
print("Recall:"+"{:.2f}%".format(recall * 100))
print("Precision:"+"{:.2f}%".format(precision * 100))
print("F1 Score:"+"{:.2f}%".format(f1_score * 100))
```

## Random Forest

```python
#Random Forest
RF_Model = RandomForestClassifier()
RF_Model.fit(X_train,y_train)
predict_rf=RF_Model.predict(X_test)
accuracy = accuracy_score(y_test,predict_rf)
recall = recall_score(y_test, predict_rf, average="weighted")
precision = precision_score(y_test, predict_rf, average="weighted")
#f1_score = f1_score(y_test, predictions, average="micro")
f1_score = 2 * ((precision * recall) / (precision + recall))

print("Random Forest")
print("Accuracy:"+"{:.2f}%".format(acccuracy * 100))
print("Recall:"+"{:.2f}%".format(recall * 100))
print("Precision:"+"{:.2f}%".format(precision * 100))
print("F1 Score:"+"{:.2f}%".format(f1_score * 100))
```

## Support Vector Machine

```python
SVM_LK = LinearSVC()
SVM_LK.fit(X_train, y_train)
#Testing the Model on test set
predict_svm=SVM_LK.predict(X_test)
acccuracy = accuracy_score(y_test, predict_svm)
recall = recall_score(y_test, predict_svm, average="weighted")
precision = precision_score(y_test, predict_svm, average="weighted")
f1_score = 2 * ((precision * recall) / (precision + recall))

print("Support Vector Machine Results")
print("Accuracy:"+"{:.2f}%".format(acccuracy * 100))
print("Recall:"+"{:.2f}%".format(recall * 100))
print("Precision:"+"{:.2f}%".format(precision * 100))
print("F1 Score:"+"{:.2f}%".format(f1_score * 100))
```

## Gradient Boosting

```python
GB_model = GradientBoostingClassifier()
GB_model.fit(X_train, y_train)
#Testing the Model on test set
predict_gb=GB_model.predict(X_test)
acccuracy = accuracy_score(y_test, predict_gb)
recall = recall_score(y_test, predict_gb, average="weighted")
precision = precision_score(y_test, predict_gb, average="weighted")
f1_score = 2 * ((precision * recall) / (precision + recall))

print("Gradient Boosting Results")
print("Accuracy:"+"{:.2f}%".format(acccuracy * 100))
print("Recall:"+"{:.2f}%".format(recall * 100))
print("Precision:"+"{:.2f}%".format(precision * 100))
print("F1 Score:"+"{:.2f}%".format(f1_score * 100))
```
